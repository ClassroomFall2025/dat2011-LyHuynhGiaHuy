```python
from NhanVien import (
    NhanVien, TiepThi, TruongPhong,
    ThemNhanVien, XoaTheoMa, TimTheoMa,
    Top5ThuNhap, HienThiDanhSach, KiemTraMaHopLe
)

def NhapSo(ThongBao):
    while True:
        GiaTri = input(ThongBao).strip()
        try:
            return float(GiaTri)
        except:
            print("Vui lòng nhập số")
def NhapChu(ThongBao):
    while True:
        GiaTri = input(ThongBao).strip()
        if GiaTri:
            return GiaTri
        print("Không được để trống")



def Menu():
    print("""
CHƯƠNG TRÌNH QUẢN LÝ NHÂN VIÊN 
1. Nhập danh sách nhân viên
2. Hiển thị danh sách nhân viên
3. Tìm nhân viên theo mã
4. Xóa nhân viên theo mã
9. Xuất 5 nhân viên có thu nhập cao nhất
0. Thoát
""")

def TaoNhanVien():
    print("Chọn loại nhân viên:")
    print("1. Nhân viên hành chính")
    print("2. Nhân viên tiếp thị")
    print("3. Trưởng phòng")

    Loai = input("Lựa chọn: ").strip()
    Ma = NhapChu("Mã nhân viên: ")

    if not KiemTraMaHopLe(Ma):
        print("Mã không hợp lệ, dùng chữ, số, gạch ngang")
        return None

    HoTen = NhapChu("Họ và tên: ")
    Luong = NhapSo("Lương cơ bản: ")

    if Loai == 1:
        return NhanVien(Ma, HoTen, Luong)
    elif Loai == 2:
        DoanhSo = NhapSo("Doanh số: ")
        TyLe = NhapSo("Tỷ lệ hoa hồng: ")
        return TiepThi(Ma, HoTen, Luong, DoanhSo, TyLe)
    elif Loai == 3:
        LuongTrachNhiem = NhapSo("Lương trách nhiệm: ")
        return TruongPhong(Ma, HoTen, Luong, LuongTrachNhiem)
    else:
        print("Lựa chọn không hợp lệ")
        return None

def Main():
    DanhSach = []

    while True:
        Menu()
        Chon = int(input("Chọn chức năng: ").strip())

        if Chon == 1:
            try:
                SoNhanVien = int(input("Nhập số nhân viên muốn thêm: "))
            except:
                print("Vui lòng nhập số nguyên")
                continue
            for i in range(SoNhanVien):
                print(f"Nhập nhân viên thứ {i+1}")
                Nv = TaoNhanVien()
                if Nv:
                    if ThemNhanVien(DanhSach, Nv):
                        print("Thêm thành công")
                    else:
                        print("Mã nhân viên đã tồn tại")
        elif Chon == 2:
            print("DANH SÁCH NHÂN VIÊN")
            HienThiDanhSach(DanhSach)
        elif Chon == 3:
            Ma = NhapChu("Nhập mã cần tìm: ")
            Nv = TimTheoMa(DanhSach, Ma)
            if Nv:
                print("Tìm thấy nhân viên:")
                print(Nv)
            else:
                print("Không tìm thấy nhân viên có mã:", Ma)
        elif Chon == 4:
            Ma = NhapChu("Nhập mã cần xóa: ")
            if XoaTheoMa(DanhSach, Ma):
                print("Xóa thành công")
            else:
                print("Không tìm thấy nhân viên có mã này")
        elif Chon == 9:
            print("Top 5 nhân viên có thu nhập cao nhất")
            Top5 = Top5ThuNhap(DanhSach)
            HienThiDanhSach(Top5)
        elif Chon == 0:
            print("Kết thúc chương trình")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn lại")
        
```
